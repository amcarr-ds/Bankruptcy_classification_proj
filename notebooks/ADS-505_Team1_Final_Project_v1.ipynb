### Label Encode target feature values {-}

```python
df01.columns = xy01_lst01
df01_ohe_fit = LabelEncoder().fit(df01['class'])
df01['class_tx'] = df01_ohe_fit.transform(df01['class'])

print(f'{df01.head(3)}\n{df01.shape}')
display(df01.info())
```

### Display overall class percentages {-}

```python
# Calc proportions by class
df01_y01_cnt = pd.value_counts(df01['class_tx'])
df01_y01_cnt_df01 = pd.DataFrame(df01_y01_cnt)
df01_y01_cnt_df01['class_percent'] = round((df01_y01_cnt_df01['class_tx'] / np.sum(df01_y01_cnt_df01['class_tx']))*100, 1)
print(f'Full data set count and percent per class:\n{df01_y01_cnt_df01}')
```

### Create train/test split {-}

```python
df02 = df01.copy()
print(df02.columns)

df02_xy01_lst01 = []

df02_x01_dum_lst01 = []

df02_x01_bin_lst01 = []

df02_x01_num_lst01 = []

df02_x01_drp_lst01 = ['class']


'''__________________________________________________________________________'''
df02 = df02.drop(df02_x01_drp_lst01, axis=1)

x01_df01 = df02.drop(['class_tx'], axis=1)
y01_df01 = df02['class_tx']


df02_null_summ01 = pd.DataFrame(df02.isnull().sum(), columns=['count'])

display(df02_null_summ01)
display(df02_null_summ01.loc[(df02_null_summ01['count'] != 0)])

'''__________________________________________________________________________'''
# Create train & test sets for X & Y (NOTE: stratification by y)
train_x01_df01, test_x01_df01, \
train_y01_df01, test_y01_df01 = train_test_split(x01_df01,
                                                 y01_df01,
                                                 test_size=.3,
                                                 random_state=1699,
                                                 stratify=y01_df01)

# Convert y df's to vector arrays
train_y01_vc01 = train_y01_df01.to_numpy()
test_y01_vc01 = test_y01_df01.to_numpy()

# Display train, val, & test set df's
print(f'\nTraining X matrix dimensions = {train_x01_df01.shape}')
print(f'Test X matrix dimensions = {test_x01_df01.shape}')

print(f'\nTraining y matrix dimensions = {train_y01_df01.shape}')
print(f'Test y matrix dimensions = {test_y01_df01.shape}')
```

### Check for and remove features with near zero variance {-}

```python
# Remove near-zero variance features
train_x01_df01_nzv_fit = VarianceThreshold().fit(train_x01_df01)
train_x02_tx_vc01 = train_x01_df01_nzv_fit.transform(train_x01_df01)

train_x02_tx_df01 = pd.DataFrame(train_x02_tx_vc01, columns=[train_x01_df01_nzv_fit.get_feature_names_out()])
test_x02_tx_df01 = test_x01_df01[train_x01_df01_nzv_fit.get_feature_names_out()]

print(f'Training X NZV transformed matrix dimensions = {train_x02_tx_df01.shape}')
print(f'Test X NZV transformed matrix dimensions = {test_x02_tx_df01.shape}')

print(f'\n{train_x01_df01.shape[1] - train_x02_tx_df01.shape[1]} near zero variance features were eliminated')
```

### Additional preprocessing (post-EDA) {-}

#### Impute missing values {-}

```python
# Implement KNN Imputer for remaining missing vals
train_x02_tx_df01_kni_fit = KNNImputer().fit(train_x02_tx_df01)
train_x02_tx_vc02 = train_x02_tx_df01_kni_fit.transform(train_x02_tx_df01)
test_x02_tx_vc02 = train_x02_tx_df01_kni_fit.transform(test_x02_tx_df01)

train_x02_tx_df02 = pd.DataFrame(train_x02_tx_vc02, columns=train_x02_tx_df01.columns)
test_x02_tx_df02 = pd.DataFrame(test_x02_tx_vc02, columns=train_x02_tx_df01.columns)

train_x02_tx_df02_null_summ01 = pd.DataFrame(train_x02_tx_df02.isnull().sum(), columns=['count'])

display(train_x02_tx_df02_null_summ01.loc[(train_x02_tx_df02_null_summ01['count'] != 0)])
```

#### Apply rebalancing technique (SMOTE) {-}

```python
# Apply SMOTE to address class imbalance-on training set only (Geeks for Geeks, 2022)
smote_txr = SMOTE(random_state=1699)
train_x02_tx_df03, train_y01_vc02 = smote_txr.fit_sample(train_x02_tx_df02, train_y01_vc01)

print(f'{train_x02_tx_df02.shape}')
print(f'{train_x02_tx_df03.head(3)}\n{train_x02_tx_df03.shape}')
print(f'{train_y01_vc02[:7]}\n{train_y01_vc02.shape}')
```

#### Scale X values {-}

```python
# Scale vals in 2 ways: z-score & min-max
train_x02_tx_df03_sts_fit = StandardScaler().fit(train_x02_tx_df03)
train_x02_tx_vc03 = train_x02_tx_df03_sts_fit.transform(train_x02_tx_df03)
test_x02_tx_vc03 = train_x02_tx_df03_sts_fit.transform(test_x02_tx_df02)

train_x02_tx_df04 = pd.DataFrame(train_x02_tx_vc03, columns=train_x02_tx_df01.columns)
test_x02_tx_df04 = pd.DataFrame(test_x02_tx_vc03, columns=train_x02_tx_df01.columns)

print(f'{train_x02_tx_df04.shape}')
print(f'{test_x02_tx_df04.shape}')

'''__________________________________________________________________________'''
train_x02_tx_df03_mms_fit = MinMaxScaler().fit(train_x02_tx_df03)
train_x02_tx_vc04 = train_x02_tx_df03_mms_fit.transform(train_x02_tx_df03)
test_x02_tx_vc04 = train_x02_tx_df03_mms_fit.transform(test_x02_tx_df02)

train_x02_tx_df05 = pd.DataFrame(train_x02_tx_vc04, columns=train_x02_tx_df01.columns)
test_x02_tx_df05 = pd.DataFrame(test_x02_tx_vc04, columns=train_x02_tx_df01.columns)

print(f'\n{train_x02_tx_df05.shape}')
print(f'{test_x02_tx_df05.shape}')
```
